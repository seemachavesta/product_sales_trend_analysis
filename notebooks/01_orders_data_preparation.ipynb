# Summary of Orders

## Orders Dataset
- Total rows: **99,441**
- Total columns: **8**
- Delivered orders: **96,478**
- Notes: Contains timestamps, order status, and customer IDs.





```python
#Filtering the order by order status
order_delivered = orders[orders['order_status'] == 'delivered'] 
```

```python
order_delivered.shape
```

```python
order_delivered.head()
```

```python
orders.columns
```

```python
order_delivered = order_delivered[[
    'order_id',
    'customer_id', 
    'order_purchase_timestamp'
]]
```

```python
order_delivered['order_purchase_timestamp'] = pd.to_datetime(
    order_delivered['order_purchase_timestamp']
)
```

```python
order_delivered.head()
```

```python
order_delivered['year'] = order_delivered['order_purchase_timestamp'].dt.year
```

```python
order_delivered.head()
```

```python
order_delivered['month'] = order_delivered['order_purchase_timestamp'].dt.month
```

```python
order_delivered['year_month'] = order_delivered['order_purchase_timestamp'].dt.to_period('M')
```

```python
order_delivered.head()
```

```python
#Converting the clean orders dataset to csv file in data_processed folder
order_delivered.to_csv("../data_processed/orders_delivered.csv", index=False)
```
